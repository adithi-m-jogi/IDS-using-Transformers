```python
import kagglehub

path = kagglehub.dataset_download("dhoogla/cicids2017")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

dataset_path = r"C:\Users\Adithi\Transformers\granules"

# ── Load all parquet files ────────────────────────────────────────
parquet_files = [f for f in os.listdir(dataset_path) if f.endswith('.parquet')]
print(f"Found {len(parquet_files)} parquet files:")
for f in parquet_files:
    print(f"  {f}")

dfs = []
for file in parquet_files:
    filepath = os.path.join(dataset_path, file)
    df_temp  = pd.read_parquet(filepath)
    print(f"Loaded {file} → {df_temp.shape}")
    dfs.append(df_temp)

df = pd.concat(dfs, ignore_index=True)
print(f"\nCombined shape: {df.shape}")

label_fix = {
    'Web Attack â Brute Force'    : 'Web Attack - Brute Force',
    'Web Attack â XSS'            : 'Web Attack - XSS',
    'Web Attack â Sql Injection'  : 'Web Attack - Sql Injection',
    'Web Attack \x96 Brute Force' : 'Web Attack - Brute Force',
    'Web Attack \x96 XSS'         : 'Web Attack - XSS',
    'Web Attack \x96 Sql Injection': 'Web Attack - Sql Injection',
}
df['Label'] = df['Label'].replace(label_fix)
df['Label'] = df['Label'].str.strip()

print(f"\nColumns: {df.columns.tolist()}")
print(f"\nLabel distribution:\n{df['Label'].value_counts()}")
```

```python
df['Label'] = df['Label'].str.replace('\ufffd', '-', regex=False)

label_rename = {
    'Web Attack - Brute Force'  : 'Web_Brute_Force',
    'Web Attack - XSS'          : 'XSS',
    'Web Attack - Sql Injection': 'SQL_Injection',
}
df['Label'] = df['Label'].replace(label_rename)

df['Label'] = df['Label'].str.replace('  ', ' ', regex=False)  # fix double spaces
df['Label'] = df['Label'].str.strip()

print(df['Label'].unique())
print(df['Label'].value_counts())
```
